```python
import requests

link = "https://api.open-meteo.com/v1/forecast?latitude=-5.4297&longitude=105.2611&daily=temperature_2m_max&timezone=Asia%2FBangkok"
respon = requests.get(link).json()

daftar_suhu = respon["daily"]["temperature_2m_max"]
daftar_hari = respon["daily"]["time"]

for i in range(len(daftar_suhu)):
    hari = daftar_hari[i]
    suhu = daftar_suhu[i]

    if suhu > 20:
        file = open("laporan_cuaca.txt", "a")
        file.write("Tanggal: " + str(hari) + " Suhu: " + str(suhu) + " Derajat\n")
        file.close()
        print("Data hari " + str(hari) + " masuk ke file")
```
