Praktik 1: Mengirim Request ke API

```python
import requests
u = "https://api.agify.io?name=budi"
res = requests.get(u)
print(res)
print(res.text)
```

Praktik 2: Menggunakan Data JSON

```python
data = res.json()
print(data)
print(data["name"])
print(data["age"])
```

Praktik 3: Input dari Pengguna

```python
n = input("masukkan nama: ")
url2 = f"https://api.agify.io?name={n}"
d2 = requests.get(url2).json()
print("Halo " + d2["name"])
print("Umur " + str(d2["age"]))
```

Praktik 4: API Quotes — Kutipan Inspiratif

```python
url_q = "https://quotes.liupurnomo.com/api/quotes/random"
d3 = requests.get(url_q).json()
print(d3["data"]["text"])
print("- " + d3["data"]["author"])
```

Praktik 5: Looping dengan API

```python
import requests
for i in range(3):
    u_loop = "https://quotes.liupurnomo.com/api/quotes?category=motivasi&page=1&limit=10"
    d4 = requests.get(u_loop).json()
    print("Quote ke-" + str(i+1))
    print(d4["data"][i+1]["text"])
```

Tugas Utama Pertemuan Ini

```python
nama_user = input("nama: ")
api_link = f"https://api.genderize.io?name={nama_user}"
final = requests.get(api_link).json()

print("HASIL PREDIKSI")
print("Nama: " + final['name'])
print("Gender: " + final['gender'])
print("Probabilitas: " + str(final['probability']))
```
